```python
# Import libraries
import pandas as pd
import os

# Define file paths and load data
data_path = "../unprocessed-data"
season_file = "nba_player_stats_2023-24.csv" # 23-24 season for intial testing
file_path = os.path.join(data_path, season_file)
season_data = pd.read_csv(file_path)

pd.set_option('display.max_columns', None)
season_data.head() # Test successful load
```

```python
# Combine season data into a single data structure
season_files = [file for file in os.listdir(data_path) if file.endswith('.csv')]
all_season_data = []

# Iterate through datasets and concatenate
for season_data_file in season_files:
    file_path = os.path.join(data_path, season_data_file)
    season_df = pd.read_csv(file_path)  # Renamed variable to avoid conflict
    all_season_data.append(season_df)

# Concatenate all the dataframes into one large dataframe
combined_data = pd.concat(all_season_data, ignore_index=True)

# Display the first/last few rows of the combined dataset
combined_data
```

```python
# Show empty values in combined data
combined_data.isnull().sum()
```

```python
# Calculate percentages from the raw values
combined_data['FG%'] = (combined_data['FG'] / combined_data['FGA']) * 100
combined_data['3P%'] = (combined_data['3P'] / combined_data['3PA']) * 100
combined_data['2P%'] = (combined_data['2P'] / combined_data['2PA']) * 100
combined_data['FT%'] = (combined_data['FT'] / combined_data['FTA']) * 100

# Calculate eFG% (Effective Field Goal Percentage)
combined_data['eFG%'] = ((combined_data['FG'] + 0.5 * combined_data['3P']) / combined_data['FGA']) * 100

# Handle division by zero (replace NaN values with 0 or another value)
combined_data['FG%'].fillna(0, inplace=True)
combined_data['3P%'].fillna(0, inplace=True)
combined_data['2P%'].fillna(0, inplace=True)
combined_data['FT%'].fillna(0, inplace=True)
combined_data['eFG%'].fillna(0, inplace=True)

# Show updated empty values in combined data
combined_data.isnull().sum()
```

```python
# Drop lines representing 'League Average'
combined_data = combined_data[combined_data['Player'] != 'League Average']

# Show updated empty values in combined data
combined_data.isnull().sum()
```

```python
# Most NaN for GS come from players in the 1980-81 season, add 0 values since data is from long ago
combined_data['GS'] = combined_data['GS'].fillna(0)

# Add 'No Award' value for 'Awards' column where Awards is NaN
combined_data.loc[combined_data['Awards'].isna(), 'Awards'] = "No Awards"

# Show updated empty values in combined data
combined_data.isnull().sum()
```

```python
# Add TS% column to the combined dataset
combined_data['TS%'] = combined_data.apply(
    lambda row: (row['PTS'] / (2 * (row['FGA'] + 0.44 * row['FTA'])) * 100)
    if (row['FGA'] + 0.44 * row['FTA']) > 0 else 0, axis=1
)

# Display the updated DataFrame
combined_data.head()
```

```python
# Add EFF column to the combined dataset
combined_data['EFF'] = combined_data.apply(
    lambda row: (
        (row['PTS'] + row['TRB'] + row['AST'] + row['STL'] + row['BLK'] -
         (row['FGA'] - row['FG']) - (row['FTA'] - row['FT']) - row['TOV']) / row['G'] * 100
    ) if row['G'] > 0 else 0, axis=1
)

# Display the updated DataFrame
combined_data.head()
```

```python
print(combined_data.columns, "\n")
# Reorder Features
columns = ['Player', 'Age', 'Team', 'Pos', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%',
           '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%',
           'TS%', 'EFF', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 
           'Awards', 'Season']

# Apply the new column order
combined_data = combined_data[columns]

# Print the columns and dataFrame
print(combined_data.columns, "\n")

combined_data.head(5)
```

```python
# Adding Team Wins, Losses, and Win/Loss % columns to each player
import pandas as pd
team_win_data = pd.read_csv('../processed-data/prepared_team_win_data.csv')
# Check
team_win_data.head()

# Modify 'Season' values in combined_data  
combined_data['Season'] = combined_data['Season'].str.split('-').str[0]
# Check
combined_data.head() 

# See all unique values in combined_data['Team'], alphabetically, and all columns in team_win_data to merge correctly
print(sorted(combined_data['Team'].unique()), "\n")

# Extract the column names, convert to list, and exclude 'Season'
team_columns = list(team_win_data.columns.difference(['Season']))
print(team_columns, '\n')

# Drop players with multiTeam values in 'Team'
combined_data = combined_data[~combined_data['Team'].isin(['2TM', '3TM', '4TM', '5TM'])]
# Change Charlotte Hornets team abbreviations (CHO, CHA = CHO) 
combined_data['Team'] = combined_data['Team'].replace({'CHH': 'CHO', 'CHA': 'CHO'})
# Change Kansas City Kings to Sacramento Kings (KCK = SAC)
combined_data['Team'] = combined_data['Team'].replace({'KCK': 'SAC'})
# Change New Jersey Nets to Brooklyn Nets (NJN = BRK)
combined_data['Team'] = combined_data['Team'].replace({'NJN': 'BRK'})
# Change New Orleans/Oklahoma City Hornets to New Orleans Pelicans (NOK, NOH = NOP)
combined_data['Team'] = combined_data['Team'].replace({'NOK': 'NOP', 'NOH': 'NOP'})
# Change San Diego Clippers to Los Angeles Clippers (SDC = LAC)
combined_data['Team'] = combined_data['Team'].replace({'SDC': 'LAC'})
# Change Seattle SuperSonics to Oklahoma City Thunder (SEA = OKC)
combined_data['Team'] = combined_data['Team'].replace({'SEA': 'OKC'})
# Change Vancouver Grizzlies to Memphis Grizzlies (VAN = MEM)
combined_data['Team'] = combined_data['Team'].replace({'VAN': 'MEM'})
# Change Washington Bullets to Washington Wizards (WSB = WAS)
combined_data['Team'] = combined_data['Team'].replace({'WSB': 'WAS'})

print(sorted(combined_data['Team'].unique()))
```

```python
# Create a temporary DataFrame with only 'Season' and the relevant team win columns from team_win_data
team_wins_only = team_win_data.set_index('Season').stack().reset_index()
team_wins_only.columns = ['Season', 'Team', 'tW']

# Ensure both 'Season' columns have the same data type (int64)
combined_data['Season'] = combined_data['Season'].astype(int)
team_wins_only['Season'] = team_wins_only['Season'].astype(int)

# Merge combined_data with team_wins_only on 'Season' and 'Team'
merged_data = pd.merge(combined_data, team_wins_only, on=['Season', 'Team'], how='left')

# Check the 5 sample rows of the result
merged_data[['Player', 'Team', 'Season', 'tW']].sample(n=5)
```

```python
# Add Team Games Played (tGP)

# Default tGP value
merged_data['tGP'] = 82

# 1998 Season / Lockout / 50 tGP
merged_data.loc[merged_data['Season'] == 1998, 'tGP'] = 50
# 2011 Season / Lockout / 66 tGP
merged_data.loc[merged_data['Season'] == 2011, 'tGP'] = 66
# 2012 Season / Boston Marathon Bombing / 81 tGP for BOS and IND
merged_data.loc[(merged_data['Season'] == 2012) & merged_data['Team'].isin(['IND', 'BOS']), 'tGP'] = 81
# 2020 Season / COVID / 72 tGP
merged_data.loc[merged_data['Season'] == 2020, 'tGP'] = 72

# 2019 Season - COVID adjustments
covid_teams = {
    64: ['MIN'],
    65: ['CHO', 'CHI', 'CLE', 'GSW'],
    66: ['NYK', 'DET'],
    67: ['ATL'],
    71: ['LAL', 'SAS'],
    72: ['TOR', 'BOS', 'BRK', 'WAS', 'LAC', 'HOU', 'OKC', 'UTA', 'SAC', 'NOP'],
    73: ['MIL', 'IND', 'MIA', 'PHI', 'ORL', 'DEN', 'MEM', 'PHO'],
    74: ['POR'],
    75: ['DAL']
}

# Apply the changes to 2019 season based on the team
for games, teams in covid_teams.items():
    merged_data.loc[(merged_data['Season'] == 2019) & (merged_data['Team'].isin(teams)), 'tGP'] = games

merged_data[merged_data['Season'] == 2019]

# Check the first few rows of the result
merged_data[['Player', 'Team', 'Season', 'tW', 'tGP']].sample(n=10)

# Sample 10 rows where Seasons are 1998, 2011, 2012, 2019, 2020
merged_data[(merged_data['Season'] == 2020) | 
            (merged_data['Season'] == 2011) |
            (merged_data['Season'] == 2012) | 
            (merged_data['Season'] == 2019) | 
            (merged_data['Season'] == 1998)][['Player', 'Team', 'Season', 'tW', 'tGP']].sample(n=10)
 
```

```python
# Add Team Games Lost (tL) and Team Win Percentage (tW%)
merged_data['tL'] = merged_data['tGP'] - merged_data['tW']
merged_data['tW%'] = (merged_data['tW'] / merged_data['tGP']).round(3)

# Remove decimals, Convert tGP, tW, and tL columns to integers
merged_data['tGP'] = merged_data['tGP'].astype(int)
merged_data['tW'] = merged_data['tW'].astype(int)
merged_data['tL'] = merged_data['tL'].astype(int)

# merged_data[['Player', 'Team', 'Season', 'tGP', 'tW', 'tL', 'tW%' ]].sample(n=10)
merged_data[(merged_data['Season'] == 2020) | 
            (merged_data['Season'] == 2011) |
            (merged_data['Season'] == 2012) | 
            (merged_data['Season'] == 2019) | 
            (merged_data['Season'] == 1998)][['Player', 'Team', 'Season', 'tGP', 'tW', 'tL', 'tW%']].sample(n=10)
```

```python
# Reorder columns before saving
merged_data = merged_data[['Player', 'Age', 'Team', 'Season', 'tGP', 'tW', 'tL', 'tW%', 'Pos', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'TS%', 'EFF', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'Awards']]
merged_data.sample(5)
```

```python
# Check for missing values in 'tGP' column
missing_values = merged_data['tGP'].isnull().sum()

# Print the result
print(f"Number of missing values in 'tGP' column: {missing_values}")
```

```python
# Code block for writing/saving processed data

# Save combined dataframe to processed data directory
processed_data_path = "../processed-data/combined_data.csv"
merged_data.to_csv(processed_data_path, index=False)
print(f"Data saved to: {processed_data_path}") # Confirmation
```
